$$
E_i(w) \equiv -l_i(w) = - y_i x_i^\top w + \text{logsumexp}(0, x_i^\top w) = - y_i z_i + \text{logsumexp}(0, z_i)
$$

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.autograd
from torch.autograd import Variable



df_iris = pd.read_csv(u'data/iris.txt',sep=' ')
df_iris

fields = ['sl','sw','pl','pw','c']

N = len(fields)
degree = 4

features_array = np.array(df_iris[fields[0:4]])
target_array = np.array(df_iris[fields[4]])
target_array[target_array>1] = 0

features = Variable(torch.from_numpy(features_array).double(),requires_grad=False)
target = Variable(torch.from_numpy(target_array).double(),requires_grad=False)
w = Variable(torch.randn(degree).double(), requires_grad=True)

eta = 0.001


def sigmoid(x):
    return 1/(1 + torch.exp(-x))
```

```python
for epoch in range(10000):

    for i in range(len(features)):
       
        f1 = torch.matmul(features[i,:], w)
        f = sigmoid(f1)
        E = -target[i]*f + torch.log(1+torch.exp(f))

        
        E.backward()
        w.data.add_(-eta * w.grad.data)

       
        w.grad.zero_()
        
    if epoch%1000 == 0: 
        print(epoch,':',E.item())
    
print(w)
    
```

```python
result1 = torch.matmul(features, w)
result = sigmoid(result1)


plt.plot(result.data.numpy().squeeze(),'o')
plt.plot(target.data.numpy().squeeze(),'.')
plt.show


#confusion_matrix = np.zeros((2,2))

#for i in range(len(result)):
#    confusion_matrix[int(target[i]), int(result[i])] += 1
    
#print(confusion_matrix)

#accuracy = np.sum(np.diag(confusion_matrix))/np.sum(confusion_matrix)

#print("Accuracy:", accuracy*100, "%")
```
