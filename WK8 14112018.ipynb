```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.autograd
from torch.autograd import Variable

x = torch.autograd.Variable(torch.linspace(-5,5,100), requires_grad = True)

y = torch.sin(x)
y1 = torch.sum(y)

y1.backward()

plt.plot(x.data.numpy(),y.data.numpy(),'o')
plt.plot(x.data.numpy(),x.grad.data.numpy(),'r')


plt.show
```

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import torch
import torch.autograd
from torch.autograd import Variable


def sigmoid(x):
    return 1/(1 + np.exp(-x))

sizes = [1,2,1]

w1 = np.random.randn(sizes[1],sizes[0])
b1 = np.random.randn(sizes[1],1)
w2 = np.random.randn(sizes[2],sizes[1])
b2 = np.random.randn(sizes[2],1)

print(w1)

def nnet(x,w1,b1,w2,b2):
    
    y1 = w1.dot(x) + b1
    f1= sigmoid(y1)
    y2 = w2.dot(y1) + b2
    f2 = sigmoid(y2)
    return f2



X = np.linspace(-50,50,100)
F = np.zeros_like(X)

for i in range(len(X)):
    F[i] = nnet(X[i],w1,b1,w2,b2)
    
plt.plot(X,F)
plt.show()
    
```
